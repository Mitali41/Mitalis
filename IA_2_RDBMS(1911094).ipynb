```python
from google.colab import files
uploaded =files.upload()
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib as plt
```

```python
import io
df=pd.read_csv(io.BytesIO(uploaded['data.csv']))
```

```python
df.head()
pre_columns = ['Name', 'Age', 'Photo', 'Nationality', 
'Overall', 'Potential', 'Club', 'Value', 'Wage','Preferred Positions'] 
Data_sheet = pd.DataFrame(data = df, columns = pre_columns) 
Top10country = Data_sheet.Nationality.value_counts() 
a = Top10country.head(10) 
a.plot(kind = 'pie', legend = True, figsize= (7,7), 
shadow= True, explode = [0,0.7,0,0,0.3,0,0,0,0,0]) 
plt.show()
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib as plt
import io
pre_columns = ['Name', 'Age', 'Photo', 'Nationality', 
'Overall', 'Potential', 'Club', 'Value', 'Wage','Preferred Positions'] 
df=pd.read_csv(io.BytesIO(uploaded['data.csv']))
df = pd.DataFrame(data = df, columns = pre_columns) 
rbao=df.groupby('Age')['Overall'].mean().reset_index()
rbao.sort_values('Age')
plt.pyplot.plot(rbao['Age'], rbao['Overall'])
plt.pyplot.show()
```
